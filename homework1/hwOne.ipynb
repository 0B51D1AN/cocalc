```python
# Imports
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("finance.csv")
```

```python
gaRows = df[df['State'] == "GEORGIA"]
```

```python
plt.figure(figsize=(10, 6))
plt.plot(gaRows.index, gaRows['Totals.Capital outlay'], marker='o')
plt.title('Totals Capital Outlay by Index in Georgia')
plt.xlabel('Index')
plt.ylabel('Totals Capital Outlay')
plt.grid()
plt.show()
```

```python
plt.figure(figsize=(10, 6))
plt.boxplot(gaRows['Totals.Capital outlay'])
plt.title('Box Plot of Totals Capital Outlay by Index in Georgia')
plt.xlabel('Index')
plt.ylabel('Totals Capital Outlay')
plt.grid()
plt.show()
```

```python
plt.figure(figsize=(10, 6))
plt.plot(gaRows.index, gaRows['Totals. Debt at end of fiscal year'], marker='o')
plt.title('Totals Debt at End of Fiscal Year by Index in Georgia')
plt.xlabel('Index')
plt.ylabel('Totals Debt at End of Fiscal Year')
plt.grid()
plt.show()
```

```python
plt.figure(figsize=(10, 6))
plt.boxplot(gaRows['Totals. Debt at end of fiscal year'])
plt.title('Box Plot of Totals. Debt at end of fiscal year in Georgia')
plt.xlabel('Georgia')
plt.ylabel('Totals Debt at end of fiscal year')
plt.grid()
plt.show()
```
